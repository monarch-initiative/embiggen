```python
import silence_tensorflow.auto
from ensmallen.datasets.linqs import Cora, get_words_data
from ensmallen.datasets.string import SpeciesTree
from embiggen.edge_prediction import GCNEdgePrediction
from embiggen.edge_label_prediction import GCNEdgeLabelPrediction
from embiggen.node_label_prediction import GCNNodeLabelPrediction
import matplotlib.pyplot as plt
import numpy as np

graph, data = get_words_data(Cora())
graph = SpeciesTree()
```

```python
model = GCNNodeLabelPrediction(
    use_simmetric_normalized_laplacian=False,
    #use_node_type_embedding=True,
    #node_feature_names="Words data",
    epochs=1
)
model.fit(graph)#, node_features=[data,])
model.plot()
```

```python
model = GCNEdgePrediction(verbose=False, epochs=10)
model.evaluate(
    number_of_holdouts=4,
    holdouts_kwargs=dict(train_size=0.8),
    evaluation_schema="Connected Monte Carlo",
    graph=graph,
    node_features=data
)
```
